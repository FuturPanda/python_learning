Exercice de graph de données

```python
dataset = {f"experience{i}" : np.random.randn(100) for i in range(4)}
dataset
def plot_dataset(dict) : 
    plt.figure(figsize=(10, 10))
    for i in range(4) :
        plt.subplot(4, 1, (i+1))
        plt.plot(dict[f"experience{i}"],dict[f"experience{i}"] )
        plt.title(f"experience {i} : ")
    
plot_dataset(dataset)
```

# Nuages de points avec dataset iris flower dataset
iris a 150 points, 4 variables, 3 classes. 
Les classes c'est ce que l'on cherche a démontrer a trouver, c'est y 
Les 4 variables, c'est ce que l'on mesure; 
On utilise y pour savoir la couleur, le rendu du bordel. 

```python
from sklearn.datasets import load_iris
iris = load_iris()
x=iris.data
y=iris.target
names = list(iris.target_names)
y
```

```python
plt.figure()
plt.scatter(x[:, 0], x[:,1], c=y, alpha=0.5, s=25)
plt.xlabel("Longueur sépal")
plt.ylabel("Largeur sépal")
```
